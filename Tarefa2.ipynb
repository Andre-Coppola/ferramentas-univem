## Importa Libs

```python
import cv2
import matplotlib.pyplot as plt 
```

## Carrega a Imagem

```python
imagem = cv2.imread("carro.jpg")
imagem2 = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

## Mostra a Imagem

```python
plt.imshow(imagem2)
plt.show()
```

## Tamanho da imagem

```python
xx=imagem.shape[0]
yy=imagem.shape[1]
print("largura:",xx,"altura:",yy)
```

## Definir valor

```python
pedaco=4
dim = int(xx /pedaco)
dim2 = int(yy /pedaco)

print ("parte_x:",dim,"parte_y:",dim2)
```

## Recortando, Salvando e Exibindo

```python
matriz = []
for x in range(pedaco): 
    for y in range(pedaco):
        print ("imagem"+str(x)+"-"+str(y)+".jpg")
        print(imagem[x*dim:(x+1)*dim, y*dim2:(y+1)*dim2].shape)
        plt.imshow(imagem[x*dim:(x+1)*dim, y*dim2:(y+1)*dim2])
        plt.show()
        cv2.imwrite("imagem"+str(x)+"-"+str(y)+".jpg",imagem[x*dim:(x+1)*dim, y*dim2:(y+1)*dim2])
   

```
